Data Preprocessing

```python
#seperating the dataset into data and label
X = wine_dataset.drop(columns = 'quality', axis = 1)
print(X)
```

Label Binarization

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
print(Y)
```

Train and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 3)
print(X.shape, X_train.shape, X_test.shape)
```

Model Training:

Random Forest Classifier

```python
model = RandomForestClassifier()
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print("Training data accuracy is: ", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print("Testing data accuracy is: ", testing_data_accuracy)
```

Building a predictive system

```python
input_data = (7.8,0.57,0.31,1.8,0.069,26.0,120.0,0.99625,3.29,0.53,9.3)
#converting list to numpy array
input_data_as_numpy = np.asarray(input_data)
#reshaping the array as we are predicting for one instance
input_data_reshape = input_data_as_numpy.reshape(1, -1)
prediction = model.predict(input_data_reshape)
print(prediction)
if prediction == 1:
  print("Quality of Wine is Good")
else:
  print("Quality of Wine is Bad")
```
